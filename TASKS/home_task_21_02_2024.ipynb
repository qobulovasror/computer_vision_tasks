```python
#  gomomorf  filtr:Battervort yuqori chastotali tasvirlarga ishlov berish
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Robert
#from scipy.integrate import quad, integrate
import numpy as np
import random
from PIL import Image, ImageDraw #Подключим необходимые библиотеки. 
import math

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np


f = Image.open("DIP3E_Original_Images_CH04/Fig0462(a)(PET_image).tif") 

fig = plt.figure(figsize=(10, 10))

rows, columns=(4,3)

fig.add_subplot(rows, columns, 1) 

plt.imshow(f, cmap='gray')
plt.axis('off')
plt.title("Berilgan tasvir")

lnn=np.log1p(np.array(f, dtype="float"))

# image in frequency domain
F = np.fft.fft2(lnn)

fig.add_subplot(rows, columns,2) 
plt.imshow(np.log1p(np.abs(F)), cmap='gray')
plt.axis('off')
plt.title("Tasvir spektri")

fig.add_subplot(rows, columns, 3) 
Fshift = np.fft.fftshift(F)
plt.imshow(np.log1p(np.abs(Fshift)), 
           cmap='gray')
plt.axis('off')
plt.title("Tasvir spektrini o`rtachalashtirish")

# Filter: Low pass filter
#M,N = f.shape

M = f.size[1] #Определяем ширину. 
N = f.size[0] #Определяем высоту. 

H = np.zeros((M,N), dtype=np.float32)
Z = np.zeros((M,N), dtype=np.float32)

D0 = 80
n=1

gammaH=2
gammaL=0.25

for u in range(M):
    for v in range(N):
        D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
        H[u,v] = (gammaH-gammaL)*(1-np.exp(-n*(D/D0)**2))+gammaL
   
        
fig.add_subplot(rows, columns, 4) 
plt.imshow(H, cmap='gray')
plt.axis('off')
plt.title("Filtr tasviri")

# Ideal Low Pass Filtering
Gshift = Fshift * H

fig.add_subplot(rows, columns, 5) 
plt.imshow(np.log1p(np.abs(Gshift)), 
           cmap='gray')
plt.axis('off')
plt.title("Tasvirga filtr  ko`paytmasi tasviri")

# Inverse Fourier Transform
G = np.fft.ifftshift(Gshift)

fig.add_subplot(rows, columns, 6) 
plt.imshow(np.log1p(np.abs(G)), 
           cmap='gray')
plt.axis('off')
plt.title("O`rtachalashtirilgan tasvir")

g = np.abs(np.fft.ifft2(G))

g = np.exp(g)

fig.add_subplot(rows, columns, 7) 
plt.imshow(g, cmap='gray')
plt.axis('off')
plt.title("Ishlov berilgan tasvir")
```

```python
#  gomomorf  filtr: Battervort yuqori chastotali tasvirlarga ishlov berish
import cv2
import numpy as np
import matplotlib.pyplot as plt

#from scipy.integrate import quad, integrate
import numpy as np
import random
from PIL import Image, ImageDraw #Подключим необходимые библиотеки. 
import math

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

f = Image.open("DIP3E_Original_Images_CH04/Fig0462(a)(PET_image).tif") 

fig = plt.figure(figsize=(10, 10))

rows, columns=(4,3)

fig.add_subplot(rows, columns, 1) 

plt.imshow(f, cmap='gray')
plt.axis('off')
plt.title("Berilgan tasvir")

F = np.fft.fft2(f)

fig.add_subplot(rows, columns,2) 
plt.imshow(np.log1p(np.abs(F)), cmap='gray')
plt.axis('off')
plt.title("Tasvir spektri")

fig.add_subplot(rows, columns, 3) 
Fshift = np.fft.fftshift(F)
plt.imshow(np.log1p(np.abs(Fshift)), 
           cmap='gray')
plt.axis('off')
plt.title("Tasvir spektrini o`rtachalashtirish")

M = f.size[1] #Определяем ширину. 
N = f.size[0] #Определяем высоту. 

H = np.zeros((M,N), dtype=np.float32)

D0 = 80
n=4
W=10

gammaH=2
gammaL=0.25

for u in range(M):
    for v in range(N):
        D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
        # H[u,v] = 1/(1+(D*W/(D**2-D0**2))**(2*n))
        H[u,v] = 1-np.exp(-1*(((D**2-D0**2)/((D+0.0000000001)*W))**2))
        
fig.add_subplot(rows, columns, 4) 
plt.imshow(H, cmap='gray')
plt.axis('off')
plt.title("Filtr tasviri")

# Ideal Low Pass Filtering
Gshift = Fshift * H

fig.add_subplot(rows, columns, 5) 
plt.imshow(np.log1p(np.abs(Gshift)), 
           cmap='gray')
plt.axis('off')
plt.title("Tasvirga filtr  ko`paytmasi tasviri")

# Inverse Fourier Transform
G = np.fft.ifftshift(Gshift)

fig.add_subplot(rows, columns, 6) 
plt.imshow(np.log1p(np.abs(G)), 
           cmap='gray')
plt.axis('off')
plt.title("O`rtachalashtirilgan tasvir")

g = np.abs(np.fft.ifft2(G))

g = np.exp(g)

fig.add_subplot(rows, columns, 7) 
plt.imshow(g, cmap='gray')
plt.axis('off')
plt.title("Ishlov berilgan tasvir")
```

### Rayleigh tasvirga shovqin qo`shish

```python
#Relya shovqini qo`shish
import numpy as np
import random
from PIL import Image, ImageDraw 
import math
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np


f = Image.open("DIP3E_Original_Images_CH04/Fig0462(a)(PET_image).tif") 

N = f.size[0] #img height
M = f.size[1] #img width. 

img_size = (M, N)
sigma = 70.0

#Generate noise
noise = np.random.rayleigh(sigma, img_size)

# add noise to img
noise_image=np.add(f, noise);


def PSNR(original, compressed): 
    mse = np.mean((original - compressed) ** 2) 
    if(mse == 0):  # MSE is zero means no noise is present in the signal . 
                  # Therefore PSNR have no importance. 
        return 100
    max_pixel = 255.0
    psnr = 20 * np.log10(max_pixel / np.sqrt(mse)) 
    return psnr
    
print(PSNR(noise_image,f))

imgnew  = Image.new( mode = "L", size = (N, M) )
imgnewpix=imgnew.load()

mediana=np.array([1,1,1,1,1,1,1,1,1], dtype=np.int32)
         
for i in range(1,N-2):
    for j in range(1, M-2):
        for s in range(3):
            for t in range(3):
                mediana[s*3+t]=noise_image[j+t,i+s]
       
        mediana=np.sort(mediana)
        imgnewpix[i,j]=int(mediana[4])


# print results
fig = plt.figure(figsize=(10, 30))
rows, columns=(1,4)


fig.add_subplot(rows, columns, 1) 
plt.imshow(f, cmap='gray')
plt.axis('off')
plt.title("Tasvir")

fig.add_subplot(rows, columns, 2) 
plt.imshow(noise, cmap='gray')
plt.axis('off')
plt.title("Shovqin")

fig.add_subplot(rows, columns, 3) 
plt.imshow(noise_image, cmap='gray')
plt.axis('off')
plt.title("Shovqinli + tasvir")

fig.add_subplot(rows, columns, 4) 
plt.imshow(imgnew, cmap='gray')
plt.axis('off')
plt.title("Mediana qo`llanilgan tasvir")

print(PSNR(np.array(imgnew),f))
```

```python
import numpy as np
import cv2 as cv
drawing = False # true if mouse is pressed
mode = True # if True, draw rectangle. Press 'm' to toggle to curve
ix,iy = -1,-1
# mouse callback function
def draw_circle(event,x,y,flags,param):
    global ix,iy,drawing,mode
    if event == cv.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
    elif event == cv.EVENT_MOUSEMOVE:
        if drawing == True:
            if mode == True:
                cv.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)
            else:
                cv.circle(img,(x,y),5,(0,0,255),-1)
    elif event == cv.EVENT_LBUTTONUP:
        drawing = False
        if mode == True:
            cv.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)
        else:
            cv.circle(img,(x,y),5,(0,0,255),-1)
            
img = np.zeros((512,512,3), np.uint8)
cv.namedWindow('image')
cv.setMouseCallback('image',draw_circle)
while(1):
    cv.imshow('image',img)
    k = cv.waitKey(1) & 0xFF
    if k == ord('m'):
        mode = not mode
    elif k == 27:
        break
cv.destroyAllWindows()
```

```python
# Gauss past chastotali tasvirlarga ishlov berish
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Robert
import random
from PIL import Image, ImageDraw #Подключим необходимые библиотеки. 
import math

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np


#mode = int(input('mode:')) #Считываем номер преобразования. 
f = Image.open("DIP3E_Original_Images_CH04/Fig0457(a)(thumb_print).tif") #Открываем изображение. 
# original image
#f = cv2.imread("/Users/alisher/Desktop/DIP3E_Original_Images_CH04/Fig0429(a)(blown_ic).tif",0)

fig = plt.figure(figsize=(10, 10))

rows, columns=(4,3)

fig.add_subplot(rows, columns, 1) 

plt.imshow(f, cmap='gray')
plt.axis('off')
plt.title("Berilgan tasvir")



# image in frequency domain
F = np.fft.fft2(f)

fig.add_subplot(rows, columns,2)

plt.imshow(np.log1p(np.abs(F)), 
           cmap='gray')
plt.axis('off')
plt.title("Tasvir spektri")

fig.add_subplot(rows, columns, 3) 
Fshift = np.fft.fftshift(F)

plt.imshow(np.log1p(np.abs(Fshift)), 
           cmap='gray')
plt.axis('off')
plt.title("Tasvir spektrini o`rtachalashtirish")

# Filter: Low pass filter
#M,N = f.shape

M = f.size[1] #Определяем ширину. 
N = f.size[0] #Определяем высоту. 


H = np.zeros((M,N), dtype=np.float32)

D0 = 50
n=1
for u in range(M):
    for v in range(N):
        D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
        H[u,v] = np.exp(-(D**2)/(2*D0**2))
        

fig.add_subplot(rows, columns, 4) 

plt.imshow(H, cmap='gray')
plt.axis('off')
plt.title("Filtr tasviri")

# Ideal Low Pass Filtering
Gshift = Fshift * H

fig.add_subplot(rows, columns, 5) 

plt.imshow(np.log1p(np.abs(Gshift)), 
           cmap='gray')
plt.axis('off')
plt.title("Tasvirga filtr  ko`paytmasi tasviri")

# Inverse Fourier Transform
G = np.fft.ifftshift(Gshift)

fig.add_subplot(rows, columns, 6) 

plt.imshow(np.log1p(np.abs(G)), 
           cmap='gray')
plt.axis('off')
plt.title("O`rtachalashtirilgan tasvir")

g = np.abs(np.fft.ifft2(G))

fig.add_subplot(rows, columns, 7)

plt.imshow(g, cmap='gray')
plt.axis('off')
plt.title("Ishlov berilgan tasvir")


# Filter: High pass filter
H = 1 - H
fig.add_subplot(rows, columns, 8) 

plt.imshow(H, cmap='gray')
plt.axis('off')
plt.title("Filtr inkor tasviri")

# Ideal High Pass Filtering
Gshift = Fshift * H

fig.add_subplot(rows, columns, 9) 

plt.imshow(np.log1p(np.abs(Gshift)), 
           cmap='gray')
plt.axis('off')
plt.title("Tasvirga filtr  ko`paytmasi tasviri")

# Inverse Fourier Transform
G = np.fft.ifftshift(Gshift)
fig.add_subplot(rows, columns, 10) 

plt.imshow(np.log1p(np.abs(G)), 
           cmap='gray')
plt.axis('off')
plt.title("O`rtachalashtirilgan tasvir")

fig.add_subplot(rows, columns, 11) 
g = np.abs(np.fft.ifft2(G))

plt.imshow(g, cmap='gray')
plt.axis('off')
plt.title("Inkor filtr bilan ishlov berilgan tasvir")
```
